## 游댌 Niveles de pureza (Clarity)

La pureza indica la cantidad y visibilidad de inclusiones internas o imperfecciones del diamante.

- **FL (Flawless)**: No presenta inclusiones ni imperfecciones visibles.
- **IF (Internally Flawless)**: Sin inclusiones internas; solo imperfecciones externas m칤nimas.
- **VVS1 / VVS2 (Very Very Slightly Included)**: Inclusiones extremadamente dif칤ciles de detectar incluso con lupa.
- **VS1 / VS2 (Very Slightly Included)**: Inclusiones peque침as, dif칤ciles de ver a simple vista.
- **SI1 / SI2 (Slightly Included)**: Inclusiones visibles con lupa y, en algunos casos, a simple vista.

## 游빍 Certificaciones de laboratorio (Report)

Los certificados garantizan la autenticidad y evaluaci칩n objetiva del diamante.

- **GIA (Gemological Institute of America)**: Laboratorio m치s prestigioso y estricto a nivel internacional.
- **IGI (International Gemological Institute)**: Muy utilizado en diamantes de laboratorio y joyer칤a comercial.
- **HRD (Hoge Raad voor Diamant)**: Referente europeo en certificaci칩n de diamantes.
- **GCAL (Gem Certification & Assurance Lab)**: Certificaci칩n premium con garant칤as adicionales.
